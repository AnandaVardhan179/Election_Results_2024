```python
import matplotlib.pyplot as plt
import numpy as np


parties = ['BJP', 'INC', 'Samajwadi', 'Aam Aadmi']
seats_2019 = [303, 52, 5, 1]
seats_2024 = [240, 99, 10, 5]

x = np.arange(len(parties))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, seats_2019, width, label='2019')
rects2 = ax.bar(x + width/2, seats_2024, width, label='2024')

.
ax.set_xlabel('Parties')
ax.set_ylabel('Number of Seats')
ax.set_title('Seats Won by Parties in Indian General Elections (2019 vs 2024)')
ax.set_xticks(x)
ax.set_xticklabels(parties)
ax.legend()

.
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
